```python
sns.lineplot(data=data, x='ModA', y='Cleaning', palette = 'Set2')
plt.title("Impact of cleaning on ModA over Time")
plt.xlabel("ModA")
plt.ylabel("Cleaning")
plt.show()
```

```python
corr_matrix=data[['GHI','DNI','DHI','TModA','TModB']].corr()
sns.heatmap(corr_matrix, annot= True, cmap = 'coolwarm')
plt.title("Correlation among solar datapoints vs. tempreture datapoints")
plt.show()
```

```python
sns.lineplot(data=data, x='RH', y='TModA', palette = 'Set2')
plt.title("Impact of relative humidity vs. TModA")
plt.xlabel("RH")
plt.ylabel("ModA")
plt.show()
```
